## Importação dos Dados


```python
import pandas as pd
from plotnine import *
```

```python
#Leitura do conjunto de dados
diamonds = pd.read_csv("diamonds.csv.gz")
```

```python
# Visualizar as primeiras linhas
print("Dimensões do conjunto de dados:", diamonds.shape)
display(diamonds.head())
```

## Gráfico 1: Relação entre peso (carat) e preço (price)

```python
grafico1 = (
    ggplot(diamonds, aes(x="carat", y="price")) +
    geom_point(alpha=0.4, color="steelblue") +
    geom_smooth(color="red") +
    labs(title='Relação entre Peso e Preço dos Diamantes',
         x='Peso (carat)', y='Preço (USD)') +
    theme_bw()
)
grafico1
```

- O que foi utilizado?

    - geom_point(alpha=0.4): para aplicar transparência aos pontos, o que ajuda a visualizar áreas com alta concentração de dados e evita o problema de sobreposição (overplotting);

    - geom_smooth(method='lm'): linha de tendência para destacar a relação positiva entre as variáveis;

    - theme_bw(), que oferece um fundo limpo e melhora a legibilidade geral.

- Achados:

    - O gráfico revela uma associação positiva entre o peso (carat) e o preço (price) dos diamantes. A linha de tendência ascendente confirma que, quanto maior o peso, maior o preço. O preço aumenta de forma muito mais acentuada para diamantes com mais quilates. Isso indica que outras características, além do peso, influenciam significativamente o valor final do diamante.

 ## Gráfico 2: Relação entre peso e preço por tipo de corte

```python
grafico2 = (
    ggplot(diamonds, aes(x='carat', y='price', color='cut')) +
    geom_point(alpha=0.4) +
    geom_smooth(method='lm', se=False) +
    labs(
        title='Relação entre Peso e Preço por Tipo de Corte',
        x='Peso (carat)',
        y='Preço (USD)',
        color='Corte'
    ) +
    theme_bw()
)
grafico2
```

- A relação positiva entre peso e preço se mantém em todos os tipos de corte, como esperado. No entanto, a visualização demonstra que essa relação não é idêntica entre as categorias. A inclinação da linha de regressão muda visivelmente conforme a qualidade do corte: para um mesmo aumento de peso (carat), o preço tende a subir mais acentuadamente nos diamantes de melhor corte (como Ideal e Premium) em comparação com os de corte inferior (Fair e Good). Isso sugere que a qualidade do corte atua como um multiplicador de valor, fazendo com que o "retorno" em preço por quilate seja maior em diamantes mais bem trabalhados.

## Gráfico 3: Relação entre preço e peso considerando corte, cor e clareza

```python
grafico3 = (
    ggplot(diamonds, aes(x="carat", y="price", color = "cut")) +
    geom_point(alpha=0.8, color="steelblue") +
    facet_grid("cut ~ color") +
    labs(
        title="Preço vs Peso por Corte e Cor do Diamante",
        x="Peso (carat)",
        y="Preço (USD)"
    ) +
    theme_bw() +
    theme(axis_text_x=element_text(rotation=45, hjust=1))
)
grafico3
```

- A principal dificuldade para criar esta visualização foi o grande número de combinações possíveis entre as três variáveis categóricas (cut, color e clarity).
A relação entre preço e peso definitivamente não é a mesma para as diferentes configurações de diamantes. O gráfico facetado demonstra que, embora a tendência positiva seja universal, a intensidade dessa relação varia drasticamente. Por exemplo, ao comparar o painel de um diamante de alta qualidade (corte Ideal e cor E) com um de qualidade inferior (corte Fair e cor J), notamos que a linha de tendência no primeiro caso é muito mais inclinada. Isso significa que o preço de um diamante de alta qualidade aumenta muito mais rapidamente com o peso do que o de um diamante de qualidade inferior, confirmando que a combinação de características premium potencializa o valor de cada quilate.
